# Implementation of the RBLP algorithm

```python
from utils import synthesize_signal_strube, calculate_coefficients, synthesize_signal_dirac
import matplotlib.pyplot as plt
```

# Sythetize excitation signals
1. Strube's glottal wave excitation (without noise)
2. Strube's glottal wave excitation (with noise)
3. Train of Dirac pulses excitation (without noise)
4. Train of Dirac pulses excitation (with noise)

```python
# AR model parameters
AR_parameters = [1, -2.22, 2.89, -3.08, 3.27, -2.77, 2.35, -1.7, 0.75]
# pitch freaquency is 125 Hz

# Synthetize signal with Strube's glottal wave excitation
signal_strube_without_noise, signal_strube_with_noise = synthesize_signal_strube(AR_parameters)

# Synthetize signal with train of Dirac pulses excitation
signal_dirac_without_noise, signal_dirac_with_noise = synthesize_signal_dirac(AR_parameters)
```

# Calculate coefficients in the case of Strube's glottal wave excitation
Case without noise

```python
A_LSQ1, A_DUTTER1, A_WLSQ1 = calculate_coefficients(signal_strube_without_noise)
plt.plot(A_LSQ1[:,0], 'b')
plt.plot(A_DUTTER1[:,0], 'r')
plt.plot(A_WLSQ1[:,0], 'k')
plt.title('$Parameter \ a_{1}$')
plt.xlabel('samples')
plt.show()
```

Case with noise

```python
A_LSQ2, A_DUTTER2, A_WLSQ2 = calculate_coefficients(signal_strube_with_noise)
plt.plot(A_LSQ2[:,0], 'b')
plt.plot(A_DUTTER2[:,0], 'r')
plt.plot(A_WLSQ2[:,0], 'k')
plt.title('$Parameter \ a_{1}$')
plt.xlabel('samples')
plt.show()
```

#  Calculate coefficients in the case of Train of Dirac pulses excitation
Case without noise

```python
A_LSQ3, A_DUTTER3, A_WLSQ3 = calculate_coefficients(signal_dirac_without_noise)
plt.plot(A_LSQ3[:,0], 'b')
plt.plot(A_DUTTER3[:,0], 'r')
plt.plot(A_WLSQ3[:,0], 'k')
plt.title('$Parameter \ a_{1}$')
plt.xlabel('samples')
plt.show()
```

```python
A_LSQ4, A_DUTTER4, A_WLSQ4 = calculate_coefficients(signal_dirac_with_noise)
plt.plot(A_LSQ4[:,0], 'b')
plt.plot(A_DUTTER4[:,0], 'r')
plt.plot(A_WLSQ4[:,0], 'k')
plt.title('$Parameter \ a_{1}$')
plt.xlabel('samples')
plt.show()
```
